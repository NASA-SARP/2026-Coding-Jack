## 2. Variables: names for values

In natural language, a **noun** is the type of that word names something: a person, a place, an
object. Programming has something similar: **variables**.

A variable is a label that points to a value. You create one with
an equals sign:

```python
x = 14
```

This tells Python: "store the number `14` in a variable called `x`."
Python figures out the **data type** (what kind of value it is) from
the value itself. Here are the four basic types:

| Data type | Python name | What it holds | Examples |
|-----------|-------------|---------------|----------|
| Integer | `int` | Whole numbers | `14`, `0`, `-3` |
| Float | `float` | Decimal numbers | `3.5`, `-0.2`, `6.022e23` |
| String | `str` | Text | `"Jack"`, `'hello'` |
| Boolean | `bool` | True or False | `True`, `False` |


> **Note:** Some programming languages make a distinction between strings (`str`) and characters (`char`). This is probably
> not important this summer, but as you get further in your programming journey it might be relevant


````python
age = 14                # int - whole number
temperature = 5.8       # float - decimal number
name = "Jack"           # str - text (needs quotes!)
is_raining = False      # bool - True or False (no quotes!)

print(age)
print(temperature)
print(name)
print(is_raining)
````

````python
# Python can tell you the type of any variable with type()
# print() is being used to tell python to actually show the output of type()
print(type(age))
print(type(temperature))
print(type(name))
print(type(is_raining))
````

````python
# If we don't use print, the output will only show the result of the last function run
# You can see here, it only shows the result of type(name)
type(age)
type(name)
````

Python knows that `True` and `False` are booleans because they are "reserved words" or "keywords". Specifically, `True` is a keyword, but `true` is not. There are many reserved words in programming languages, and they are not identical between all languages.

Look what happens when you get the capitalization wrong:


````python
print(type("True"))   # This is a str, notice the quotes!
print(type(True))     # This is a bool, no quotes!
````

## 3. Comparisons and booleans

Now that we have variables, we can ask the computer questions.
Questions that have a yes-or-no answer produce a **boolean** value: either `True` or `False`.

We ask questions using **comparison operators**:

| Operator | Meaning | Example | Result |
|----------|---------|---------|--------|
| `==` | Equal to | `4 == 4` | `True` |
| `!=` | Not equal to | `4 != 5` | `True` |
| `>` | Greater than | `5 > 3` | `True` |
| `<` | Less than | `5 < 3` | `False` |
| `>=` | Greater or equal | `5 >= 5` | `True` |
| `<=` | Less or equal | `5 <= 3` | `False` |


> **Watch out!** `==` (two equals signs) is a *comparison*. `=` (one
> equals sign) is *assignment*. Mixing them up is a common bug.

> **Note!** `!=` is the most common "not equal" operator, but some
> languages (MATLAB) use `~=` for "not equal"

````python
# Run these comparisons, what do you expect to see?
print(4 + 5 == 10)       # Is 4+5 equal to 10?
print(2**10 >= 1000)     # Is 2^10 greater-or-equal to 1000?
print(6 != 3 + 3)        # Is 6 NOT equal to 3+3?

# Comparing strings works too
state_code = 'WI'
print(state_code == 'NY')   # Is the state code New York?


# Note!! Some languages do not allow the use of `==` to compare strings. 
# These languages usually have a specific function for string comparisons
````

### Checking multiple conditions: `and` and `or`

Sometimes you need to check more than one thing at once:

- `and` - **both** sides must be true
- `or` - at least **one** side must be true

| Expression | Result |
|------------|--------|
| `True and True` | `True` |
| `True and False` | `False` |
| `True or False` | `True` |
| `False or False` | `False` |



````python
print(4 == 4 and 5 == 6)    # True and False → ?
print(5 < 6 and 6 < 7)      # True and True → ?
print(4 == 4 or 5 == 6)     # True or False → ?
````

## 4. Making decisions with `if` statements

Code normally runs from top to bottom, one line after another, just like how you would read 
paragraph. But sometimes you need the computer to **choose**
what to do based on a condition. That's what `if` statements are for.

### The structure

```python
if <condition that is True or False>:
    <do this>
    <indented block>
elif <another condition>:
    <do this instead>
else:
    <none of the above, do this>
```

The colon (`:`), the indentation, and `elif`/`else` are all required
syntax. Python uses indentation (spaces at the start of a line) to
tell which lines belong inside the `if` statement.

> Python checks conditions **top to bottom** and exits the `if`
> statement as soon as it finds the first one that's `True`.

> Python doesn't require the `end` keyword to close loops, but other languages (`bash`, `julia`, `MATLAB`) do require an `end` keyword.


> Same thing applies to the colon `:`


````python
# Should we process this satellite image?
cloud_cover = 50

if cloud_cover < 30:
    print('Good image for processing')
else:
    print('Too many clouds for processing')
````

````python
# Multiple conditions with elif
cloud_cover = 35

if cloud_cover < 30:
    print('Good image for processing')
elif cloud_cover < 40:
    print('Potentially usable')
elif cloud_cover < 50:
    print('Only if we are really desperate')
else:
    print('Too many clouds for processing')

# Try changing cloud_cover to 25, 45, 70 and re-running!
````

## 5. Lists: a group of values

So far every variable held one thing. But real data almost always
comes in groups: a collection of temperature readings, a set of
filenames, a list of station IDs.

A **list** is a data structure that holds multiple values. You write
it with square brackets `[]`, and separate items with commas:

```python
pollutants = ['co', 'co2', 'no2', 'o3']
```

Lists can hold any type: strings, numbers, booleans, or even a mix of multiple types!


````python
pollutants = ['co', 'co2', 'no2', 'o3']
ph_values = [4.5, 1.0, 4.3, 4.3, 7.6]
mixed = ['abc', 34, True, 40, 'male']

print(pollutants)
print(type(pollutants))
print(len(pollutants))    # How many items?
````

### Accessing items: indexing

You grab an item from a list by its **index**: a number that says
where it sits in the list. And here's the first Big Python Rule:

> **Python counts from 0, not from 1.**

So `list[0]` is the first item. `list[1]` is the second. The last
item is `list[-1]`. This is not the same in every language. Some languages index the last items as `list[end]`.
You can grab a slice (a range) with a colon: `list[start:end]`.
The slice goes from `start` up to (but not including) `end`.

| Expression | What you get |
|------------|-------------|
| `colors[0]` | First item |
| `colors[1]` | Second item |
| `colors[-1]` | Last item |
| `colors[1:3]` | Items at index 1 and 2 |



````python
colors = ['red', 'green', 'blue', 'yellow', 'white', 'black']

print(colors[0])       # First
print(colors[1])       # Second
print(colors[-1])      # Last
print(colors[1:3])     # Slice: indices 1 and 2
print(colors[:3])      # From start up to (not including) index 3
print(colors[3:])      # From index 3 to the end
````

### Doing things to lists: methods

If variables are like nouns, **methods** are like verbs, they let you
*do* something to an object. Two common ones for lists:

- `len(list)` - returns the number of items
- `list.append(value)` - adds a value to the end
- `list[index] = new_value` -changes the item at that position


````python
pollutants = ['co', 'co2', 'no2', 'o3']
print('Before:', pollutants)

pollutants.append('ch4')      # Add methane to the list
pollutants[2] = 'bc'          # Change 'no2' to 'bc' (black carbon)

print('After: ', pollutants)
print('Length:', len(pollutants))
````

## 6. Dictionaries: named values

A list gives you items by *position* (index 0, index 1, ...). Sometimes
that's not enough, sometimes you want to find things by *name*.

A **dictionary** stores **key-value pairs**. You use curly braces `{}`:

```python
metadata = {
    'location': 'Nairobi',
    'quality_flag': 5,
    'zenith': 60,
    'clouds': True,
}
```

Think of it like a real dictionary: the **key** is the word you look
up, and the **value** is the definition. Access the value with square
brackets: `metadata['location']`.


````python
metadata = {
    'location': 'Nairobi',
    'quality_flag': 5,
    'zenith': 60,
    'clouds': True,
}

print(metadata['location'])
print(metadata['zenith'])
````

### Inspecting and editing dictionaries

- `.keys()` - see all the keys
- `.values()` - see all the values
- `dict['key'] = value` - add a new key or overwrite an existing one


````python
metadata = {
    'location': 'Nairobi',
    'quality_flag': 5,
    'zenith': 60,
    'clouds': True,
}

# Add a new key-value pair
metadata['agency'] = 'NASA'

# Overwrite an existing value
metadata['zenith'] = 30

print('Keys:  ', list(metadata.keys()))
print('Values:', list(metadata.values()))
````

## 7. For loops: doing the same thing to many items

One of the most common tasks in data work is: *"do something to
every file in this list."* Loops let you repeat code for each
item in a collection.

The syntax is straightforward:

```python
for <variable> in <list>:
    <indented code, run once per item>
```

The loop variable (you pick the name) takes on each value in the
list, one at a time. The indented block runs once for each item.

> **The use case you'll see all the time:** looping over a list of
> filenames to read, process, or check each one.


````python
file_list = ['data_jan.csv', 'data_feb.csv', 'data_mar.csv', 'data_apr.csv']

for filename in file_list:
    print(f"Processing {filename}...")
print("Done!")     # Not indented, runs only once, after the loop
````

````python
# You can put an if statement inside a loop
file_list = ['data_jan.csv', 'data_feb.csv', 'data_mar.csv', 'data_apr.csv']

for filename in file_list:
    if 'jan' in filename:
        print(f"📅 {filename}: January data found")
    else:
        print(f"    {filename}: Other month")
````
